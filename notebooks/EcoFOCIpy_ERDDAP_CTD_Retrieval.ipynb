# EcoFOCIpy simple ERDDAP example

The simple erddap wrapper here is to accelerate the retrieval of a single profile from a cruise.  You will need to already know:
- url to the erddap service (http://akutan.pmel.noaa.gov:8080/erddap or http://ecofoci-field.pmel.noaa.gov:8080/erddap is the internal resource
- cruiseid (e.g. dy2103)
- consecutive cast number (e.g. ctd001)
- qc level (final or preliminary)

Much of this above requirements can be found by exploring the web interface for this erddap

```python
import EcoFOCIpy.io.erddap as erddap
```

```python
url = "http://akutan.pmel.noaa.gov:8080/erddap"
```

```python
erddap.test_erddap_connection(url=url) #basic test for connection to internal server, no response is a good thing
```

```python
erddap.erddapCTDretrieve(url=url,cruiseid='dy1908',qclevel='final')
```
